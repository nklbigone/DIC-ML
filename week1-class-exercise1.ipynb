## 【Problem 1] Number of grains of rice on the 100th day

```python
day = 100
grain = 1
list_grains, total_grains = [1],[1]
for i in range(1,day):
    grain += grain
    list_grains.append(list_grains[-1]*2)
    total_grains.append(sum(list_grains))
print ("Grains of rice you receive in total on the 100th day", grain)
```

```python
import  matplotlib.pyplot  as plt
%matplotlib inline 
plt.title( "Graph of Rice at 100th day " )
plt.xlabel( "Number of days " )
plt.ylabel( " Grains " )
plt.plot (list_grains, color = 'green', label='Daily')
plt.plot (total_grains, color = 'red', label='Total')
plt.legend()
plt.show()
```

## 【Problem 2] Number of rice grains outside of the 100th day

```python
def compute_sorori_shinzaemon(day = 100):
    list_grains, total_grains = [1],[1]
    for i in range(day):
        list_grains.append(list_grains[-1]*2)
        total_grains.append(sum(list_grains))
    return list_grains, total_grains
```

```python
import  matplotlib.pyplot  as plt
%matplotlib inline 
list_grains, total_grains = compute_sorori_shinzaemon(5)
plt.title( " Graph of Rice at 5th day " )
plt.xlabel( "Number of days " )
plt.ylabel( " Grains " )
plt.plot (list_grains, color = 'green', label='Daily')
plt.plot (total_grains, color = 'red', label='Total')
plt.legend()
plt.show()
```

## 【Problem 3] How many people can live for how many days with the rice we get?

```python
def survive_days(grains,people):
    grains_per_day = 200 
    days = (grains/people)/grains_per_day
    return days
print(survive_days(9005600,5),"Days")
```

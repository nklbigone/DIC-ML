## [Problem 1] Creating random numbers

```python
import numpy as np
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
seed = np.random.seed(0)
random_number = np.random.multivariate_normal(mean, cov, 500)
print(random_number)
random_number.shape
```

## [Problem 2] Visualization by scatter plot

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.title("Visualization by scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(random_number[:,0], random_number[:,1], marker='^')
plt.show()
```

## [Problem 3] Visualization by histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('x1')
plt.ylabel('Frequency')
plt.title('Histogram of x1')
plt.hist(random_number[:,0], bins=80)
plt.xlim(left=-7)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('x2')
plt.ylabel('frequency')
plt.title('Histogram of x2')
plt.hist(random_number[:,1], bins=80)
plt.xlim(right=5)
plt.show()
```

## [Problem 4] Addition of data

```python
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
seed = np.random.seed(0)
random_number2 = np.random.multivariate_normal(mean, cov, 500)
print(random_number2)
random_number2.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.title(" scatter ")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(random_number[:,0], random_number[:,1], label='0', marker='o', color='blue')
plt.scatter(random_number2[:,1], random_number2[:,0], label='1', marker='^',color='green')
plt.legend()
plt.show()
```

## [Problem 5] Data combination

```python
data_combine = np.concatenate((random_number, random_number2),axis=0)
print(data_combine)
data_combine.shape
```

```python
zeros = np.zeros(500)
ones = np.ones(500)
added_column = np.concatenate((zeros, ones))
new_array = np.append(data_combine, added_column[:,None],1)
print(new_array)
new_array.shape
```
